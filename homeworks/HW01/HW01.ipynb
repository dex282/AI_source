```python
## Домашнее задание HW01: Знакомство с NumPy
# 1. Импорт библиотеки NumPy с общепринятым сокращением
import numpy as np

#  2. Примеры создания массивов 
# 2.1. Создание одномерного массива из обычного списка
list_data = [15, 30, 65, 75, 85]
array_list = np.array(list_data)
print("1. Массив из списка:", array_list)

# 2.2. Создание массива с использованием zeros
array_zeros = np.zeros((2, 3))
print("\n2. Массив из нулей (2x3):\n", array_zeros)

# 2.3. Создание массива с использованием arange
# Создание массива из чисел от 0 до 14 с шагом 3
array_arange = np.arange(0, 15, 3)
print("\n3. Массив с arange:", array_arange)

# 3. Демонстрация атрибутов 
# Вывод формы (shape) и типа данных (dtype) для массива array_list
print("\n--- Атрибуты array_list ---")
print("Форма (shape):", array_list.shape) # Должно быть (5,)
print("Тип данных (dtype):", array_list.dtype) # Должно быть int64

#  4. Примеры индексирования и срезов 
# 4.1. Получение отдельного элемента
element = array_list[3]
print("\n4. Отдельный элемент (array_list[3]):", element)

# 4.2. Извлечение подмассива с помощью среза
# Извлечение элемента с индекса 1 по 4 
subarray = array_list[1:4]
print("5. Подмассив (срез 1:4):", subarray)

#  5. Примеры векторных операций и базовых статистик 
# 5.1. Поэлементное умножение массивов (Векторная операция)
A = np.array([2, 4, 6])
B = np.array([10, 5, 2])
C = A * B # Поэлементное умножение: 2*10, 4*5, 6*2
print("\n6. Поэлементное умножение A * B:", C)

# 5.2. Вычисление средней величины (mean) 
# Вычисление средних значений элементов массива array_list
average_value = np.mean(array_list)
print("7. Среднее значение (mean) array_list:", average_value)
```
